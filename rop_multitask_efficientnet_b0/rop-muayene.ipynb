```python
# Yollar
CSV_PATH = r"C:\Users\Seyid\Downloads\archive (1)\infant_retinal_database_info.csv"
IMG_DIR  = r"C:\Users\Seyid\Downloads\archive (1)\images_stack_without_captions\images_stack_without_captions"

# CSV
df = pd.read_csv(CSV_PATH, sep=";")
# PA float gelebiliyor  int yap
df["POSTCONCEPTUAL AGE (PA)"] = df["POSTCONCEPTUAL AGE (PA)"].astype(int)

# Dosyalar listele
img_files = [f for f in os.listdir(IMG_DIR) if f.lower().endswith((".jpg",".jpeg",".png"))]

# Dosya ad kalb: 019_M_GA35_BW3050_PA46_DG11_PF0_D1_S03_3.jpg
pat = re.compile(
    r"^(?P<ID>\d+)_"
    r"(?P<SEX>[MF])_"
    r"GA(?P<GA>\d+)_"
    r"BW(?P<BW>\d+)_"
    r"PA(?P<PA>\d+)_"
    r"DG(?P<DG>-?\d+)_"
    r"PF(?P<PF>-?\d+)_"
    r"D(?P<D>\d+)_"
    r"S(?P<S>\d+)_"
    r"(?P<FRAME>\d+)\.(?:jpg|jpeg|png)$",
    re.IGNORECASE
)

parsed = []
bad_names = []
for f in img_files:
    m = pat.match(f)
    if not m:
        bad_names.append(f); 
        continue
    g = m.groupdict()
    parsed.append({
        "FILENAME": f,
        "ID": int(g["ID"]),
        "SEX": g["SEX"],
        "GA": int(g["GA"]),
        "BW": int(g["BW"]),
        "PA": int(g["PA"]),
        "DG": int(g["DG"]),
        "PF": int(g["PF"]),
        "DEVICE": int(g["D"]),
        "SERIE": int(g["S"]),
        "FRAME": int(g["FRAME"])
    })

df_imgs = pd.DataFrame(parsed)
print("Parse edilen grsel says:", len(df_imgs))
print("Patterne uymayan isim says:", len(bad_names))
df_imgs.head()
```

```python
# Eletirme
merged = pd.merge(
    df_imgs, df,
    left_on=["ID","SEX","GA","BW","PA","DG","PF","DEVICE","SERIE"],
    right_on=["ID","SEX","GESTATIONAL AGE (GA)","BIRTH WEIGHT (BW)","POSTCONCEPTUAL AGE (PA)",
              "DIAGNOSIS CODE (DG)","PLUS FORM (PF)","DEVICE (D)","SERIE NUMBER (S)"],
    how="inner"
)

print("Toplam eleen kayt:", len(merged))

# Ka dosya elemedi?
matched_files = set(merged["FILENAME"])
unmatched = [f for f in img_files if f not in matched_files]
print("Elemeyen grsel says:", len(unmatched))
```

```python
# Etiket snflar
dg_classes = sorted(merged["DIAGNOSIS CODE (DG)"].unique().tolist())
pf_classes = sorted(merged["PLUS FORM (PF)"].unique().tolist())

dg2idx = {c:i for i,c in enumerate(dg_classes)}
pf2idx = {c:i for i,c in enumerate(pf_classes)}

merged["dg_idx"] = merged["DIAGNOSIS CODE (DG)"].map(dg2idx)
merged["pf_idx"] = merged["PLUS FORM (PF)"].map(pf2idx)

# Grup tablosu (szntya kar ayn serinin tm frameleri birlikte kalsn)
group_cols = ["ID","DEVICE (D)","SERIE NUMBER (S)"]
grp = (merged
       .groupby(group_cols, as_index=False)
       .agg({
           "FILENAME":"count",
           "dg_idx":"first",
           "pf_idx":"first",
           "DEVICE (D)":"first"
       })
       .rename(columns={"FILENAME":"N"})
      )

# Stratify etiketi: cihaz+PF (gerekirse DG'yi de kat)
grp["strat"] = grp["DEVICE (D)"].astype(str) + "|" + grp["pf_idx"].astype(str)

# Train/Val/Test blme (grup indeksleri stnden)
train_grp_idx, test_grp_idx = train_test_split(
    grp.index.values, test_size=0.2, random_state=42, stratify=grp["strat"]
)
# train  train/val
sub = grp.iloc[train_grp_idx]
train_grp_idx2, val_grp_idx = train_test_split(
    sub.index.values, test_size=0.1, random_state=42, stratify=sub["strat"]
)

# Hangi gruplar?
train_groups = set(grp.loc[train_grp_idx2, ["ID","DEVICE (D)","SERIE NUMBER (S)"]].itertuples(index=False, name=None))
val_groups   = set(grp.loc[val_grp_idx,   ["ID","DEVICE (D)","SERIE NUMBER (S)"]].itertuples(index=False, name=None))
test_groups  = set(grp.loc[test_grp_idx,  ["ID","DEVICE (D)","SERIE NUMBER (S)"]].itertuples(index=False, name=None))

def mask_from_groups(dfm, groups):
    keys = list(zip(dfm["ID"], dfm["DEVICE (D)"], dfm["SERIE NUMBER (S)"]))
    return np.array([k in groups for k in keys])

train_mask = mask_from_groups(merged, train_groups)
val_mask   = mask_from_groups(merged, val_groups)
test_mask  = mask_from_groups(merged, test_groups)

train_df = merged[train_mask].reset_index(drop=True)
val_df   = merged[val_mask].reset_index(drop=True)
test_df  = merged[test_mask].reset_index(drop=True)

len(train_df), len(val_df), len(test_df)
```

```python
IM_SIZE = 224
BATCH_SIZE = 32
NUM_WORKERS = 2

# ImageNet normalization
IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD  = [0.229, 0.224, 0.225]

train_tfms = transforms.Compose([
    transforms.RandomResizedCrop(IM_SIZE, scale=(0.8, 1.0)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.1, hue=0.05),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

eval_tfms = transforms.Compose([
    transforms.Resize(int(IM_SIZE*1.15)),
    transforms.CenterCrop(IM_SIZE),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
])

class ROPDataset(Dataset):
    def __init__(self, dfm, img_dir, tfms):
        self.dfm = dfm.copy()
        self.img_dir = img_dir
        self.tfms = tfms

    def __len__(self): return len(self.dfm)

    def __getitem__(self, i):
        row = self.dfm.iloc[i]
        path = os.path.join(self.img_dir, row["FILENAME"])
        with Image.open(path) as img:
            img = img.convert("RGB")
            img = self.tfms(img)
        dg = int(row["dg_idx"])
        pf = int(row["pf_idx"])
        device_id = int(row["DEVICE (D)"])
        return img, dg, pf, device_id

train_ds = ROPDataset(train_df, IMG_DIR, train_tfms)
val_ds   = ROPDataset(val_df,   IMG_DIR, eval_tfms)
test_ds  = ROPDataset(test_df,  IMG_DIR, eval_tfms)

# Snf arlklar + Sampler (train)
dg_counts = train_df["dg_idx"].value_counts().sort_index().values
pf_counts = train_df["pf_idx"].value_counts().sort_index().values

dg_w = 1.0 / (dg_counts + 1e-9)
pf_w = 1.0 / (pf_counts + 1e-9)

# rnek bana arlk = DG ve PF arlklarnn ortalamas
sample_weights = []
for _, r in train_df.iterrows():
    w = 0.5 * dg_w[r["dg_idx"]] + 0.5 * pf_w[r["pf_idx"]]
    sample_weights.append(w)
sample_weights = torch.DoubleTensor(sample_weights)

sampler = WeightedRandomSampler(sample_weights, num_samples=len(sample_weights), replacement=True)

train_loader = DataLoader(train_ds, batch_size=BATCH_SIZE, sampler=sampler,
                          num_workers=0, pin_memory=True)
val_loader   = DataLoader(val_ds, batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=0, pin_memory=True)
test_loader  = DataLoader(test_ds, batch_size=BATCH_SIZE, shuffle=False,
                          num_workers=0, pin_memory=True)


len(train_ds), len(val_ds), len(test_ds)
```

```python
num_dg = len(dg_classes)
num_pf = len(pf_classes)

class ROPNet(nn.Module):
    def __init__(self, num_dg, num_pf):
        super().__init__()
        # Pretrained backbone
        try:
            weights = models.EfficientNet_B0_Weights.IMAGENET1K_V1
            base = models.efficientnet_b0(weights=weights)
            feat_dim = base.classifier[1].in_features
        except:
            base = models.efficientnet_b0(pretrained=True)
            feat_dim = base.classifier[1].in_features

        self.backbone = base.features
        self.pool = nn.AdaptiveAvgPool2d(1)
        self.dropout = nn.Dropout(0.2)
        self.dg_head = nn.Linear(feat_dim, num_dg)
        self.pf_head = nn.Linear(feat_dim, num_pf)

    def forward(self, x):
        x = self.backbone(x)
        x = self.pool(x).flatten(1)
        x = self.dropout(x)
        dg_logits = self.dg_head(x)
        pf_logits = self.pf_head(x)
        return dg_logits, pf_logits

model = ROPNet(num_dg, num_pf).to(device)

# Class weights (opsiyonel, sampler zaten var; yine de ekleyebiliriz)
dg_class_weights = torch.tensor(dg_w, dtype=torch.float32, device=device)
pf_class_weights = torch.tensor(pf_w, dtype=torch.float32, device=device)

loss_dg = nn.CrossEntropyLoss(weight=dg_class_weights)
loss_pf = nn.CrossEntropyLoss(weight=pf_class_weights)

optimizer = torch.optim.AdamW(model.parameters(), lr=2e-4, weight_decay=1e-4)
scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=10)
scaler = GradScaler()
```

```python
#  IMPORTLAR
import torch
import torch.nn as nn
import torch.optim as optim
from torch.cuda.amp import autocast, GradScaler
from torch.utils.data import DataLoader, Dataset
import numpy as np
from sklearn.metrics import accuracy_score, f1_score
from tqdm.notebook import tqdm

#  DEVICE
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#  MODEL, LOSS, OPTIMIZER, SCALER, SCHEDULER (rnek)
model = model  # kendi modelini buraya koy
model.to(device)

loss_dg = nn.CrossEntropyLoss()
loss_pf = nn.CrossEntropyLoss()

optimizer = optim.Adam(model.parameters(), lr=1e-4)
scaler = GradScaler()
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=1, gamma=0.95)

#  DATA LOADERS
# train_loader ve val_loader nceden tanmlanmal
# rnek:
# train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
# val_loader = DataLoader(val_dataset, batch_size=32, shuffle=False)

#  EKSTRA BLGLER
dg_classes = ["class0", "class1"]    # rnek
pf_classes = ["class0", "class1", "class2"]
dg2idx = {c:i for i,c in enumerate(dg_classes)}
pf2idx = {c:i for i,c in enumerate(pf_classes)}
IM_SIZE = (224,224)

#  RUN EPOCH FONKSYONU
def run_epoch(loader, model, optimizer=None, scaler=None):
    is_train = optimizer is not None
    model.train() if is_train else model.eval()

    total_loss = 0.0
    all_dg_true, all_dg_pred = [], []
    all_pf_true, all_pf_pred = [], []

    for imgs, dg, pf, dev in tqdm(loader, desc="Batches"):
        imgs = imgs.to(device, non_blocking=True)
        dg = dg.to(device, non_blocking=True)
        pf = pf.to(device, non_blocking=True)

        with autocast(enabled=True):
            dg_logits, pf_logits = model(imgs)
            ldg = loss_dg(dg_logits, dg)
            lpf = loss_pf(pf_logits, pf)
            loss = ldg + lpf

        if is_train:
            optimizer.zero_grad(set_to_none=True)
            scaler.scale(loss).backward()
            scaler.step(optimizer)
            scaler.update()

        total_loss += loss.item() * imgs.size(0)

        # preds
        dg_pred = torch.argmax(dg_logits, dim=1)
        pf_pred = torch.argmax(pf_logits, dim=1)

        all_dg_true.extend(dg.detach().cpu().numpy().tolist())
        all_dg_pred.extend(dg_pred.detach().cpu().numpy().tolist())
        all_pf_true.extend(pf.detach().cpu().numpy().tolist())
        all_pf_pred.extend(pf_pred.detach().cpu().numpy().tolist())

    n = len(loader.dataset)
    avg_loss = total_loss / max(n,1)

    dg_acc = accuracy_score(all_dg_true, all_dg_pred)
    pf_acc = accuracy_score(all_pf_true, all_pf_pred)
    dg_f1  = f1_score(all_dg_true, all_dg_pred, average="macro")
    pf_f1  = f1_score(all_pf_true, all_pf_pred, average="macro")

    return {
        "loss": avg_loss,
        "dg_acc": dg_acc, "dg_f1": dg_f1,
        "pf_acc": pf_acc, "pf_f1": pf_f1,
        "dg_true": np.array(all_dg_true), "dg_pred": np.array(all_dg_pred),
        "pf_true": np.array(all_pf_true), "pf_pred": np.array(all_pf_pred),
    }

#  ETM DNGS
EPOCHS = 10
best_val = None
best_path = "rop_multitask_efficientnet_b0.pt"

for epoch in range(1, EPOCHS+1):
    print(f"\n=== Epoch {epoch} ===", flush=True)
    
    # Train
    tr = run_epoch(train_loader, model, optimizer, scaler)

    # Validation
    with torch.no_grad():
        va = run_epoch(val_loader, model, optimizer=None)

    # Scheduler step
    scheduler.step()

    # Epoch zeti
    print(f"Train Loss {tr['loss']:.4f} | DG acc {tr['dg_acc']:.3f} f1 {tr['dg_f1']:.3f} | "
          f"PF acc {tr['pf_acc']:.3f} f1 {tr['pf_f1']:.3f}", flush=True)
    print(f"Val Loss {va['loss']:.4f} | DG acc {va['dg_acc']:.3f} f1 {va['dg_f1']:.3f} | "
          f"PF acc {va['pf_acc']:.3f} f1 {va['pf_f1']:.3f}", flush=True)

    # En iyi modeli kaydet
    score = va["dg_f1"] + va["pf_f1"]
    if (best_val is None) or (score > best_val):
        best_val = score
        torch.save({
            "model_state": model.state_dict(),
            "dg_classes": dg_classes,
            "pf_classes": pf_classes,
            "dg2idx": dg2idx, "pf2idx": pf2idx,
            "img_size": IM_SIZE
        }, best_path)
        print(f" En iyi model kaydedildi  {best_path}", flush=True)
```
